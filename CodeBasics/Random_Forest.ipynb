# Random Forest

```python
# Random Forest is a machine Learning algorithm used for both Regression and Classification
```

```python
# Importing the digits dataset

import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
# Checking what the data contain

dir(digits)
```

```python
# Visualizing the data

import matplotlib.pyplot as plt
%matplotlib inline

plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
# Creating the dataframe of the data

df = pd.DataFrame(digits['data'])
df.head()
```

```python
# Appending the target column in the dataframe

df['target'] = digits['target']
df.head()
```

```python
# Splitting the data into training and testing

x = df.drop(['target'],axis=1)
y = df['target']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
# Creating the model object

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()

# ensemble is used when we used multiple algorithms to predict the outcome (We are making multiple decision tress
# take the majority vote as a outcome)

# RandomForestClassifier(n_estimators=50) is used for number of trees
```

```python
# Training the model

model.fit(x_train,y_train)
```

```python
# Making Predictions

y_predicted = model.predict(x_test)
y_predicted
```

```python
# Calculating the accuracy of the test data set

model.score(x_test,y_test)
```

```python
# Calculating the accuracy of the train data set

model.score(x_train,y_train)
```

```python
# Making the Confusion Matrix

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
# Using Seaborn for plotting Confusion Matrix

import seaborn as sn

plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
