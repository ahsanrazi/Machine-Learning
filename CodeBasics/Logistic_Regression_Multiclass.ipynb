# Logistic Regression (Multiclass Classification)

```python
# Identify Hand Written digits Recognition. 

# In this we are using the sklearn predefined datasets.
```

```python
# Importing Libraries
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_digits
```

```python
# Loading the training set

digits = load_digits()

# This data set contain digits of size 8x8
```

```python
# Check what the training set contain

dir(digits)

# data has numeric-data of 8x8
# images has digits images
# target has actual digits in images
```

```python
# printing the element

digits.data[0]

# As the digit is of 8x8 So this is the array of (8x8 = 64) 
```

```python
# plotting the digits

plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
# Printing the target

digits.target[0:5]
```

```python
# We use (data) and (target) to train our model
```

```python
# Splitting the data into training and testing

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

```python
len(x_train)
```

```python
# Creating the model object

from sklearn.linear_model import LogisticRegression

# model = LogisticRegression()
model = LogisticRegression(max_iter=500)
```

```python
# Training the model

model.fit(x_train,y_train)
```

```python
# Making Predictions

plt.matshow(digits.images[67])
```

```python
digits.target[67]
```

```python
model.predict([digits.data[67]])
```

```python
model.predict(digits.data[0:5])
```

```python
# Saving the predicted values in variale for confusion matrix

y_predicted = model.predict(x_test)
```

```python
# Calculating score of the test data

model.score(x_test,y_test)
```

```python
# Calculating score of the train data

model.score(x_train,y_train)
```

### Confusion Matrix

```python
# Confusion Matrix hepls us to get the overall feelings of our model accuracy.
```

```python
# Import Confusion Matrix

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predicted)          # confusion_matrix(actual,predicted)
cm
```

```python
# Using Seaborn for plotting Confusion Matrix

import seaborn as sn

plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
