# Training and Testing Data Split

```python
# Split available dataset into training and test using sklearn.
```

```python
# When we have a dataset then train the model on the entire dataset is not a good strategy So we have to split
# the dataset into train and test data in order to test the model on the data that is not seen before.

# Split the dataset by the 80/20 rule 80% for training and 20% for testing.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data_Set\\Train_Test\\carprices.csv')
df
```

```python
%matplotlib inline

plt.title('Milage VS Sell_Price')
plt.xlabel('Milage')
plt.ylabel('Sell Price $')
plt.scatter(df['Mileage'], df['Sell Price($)'])
```

```python
plt.title('Age VS Sell_Price')
plt.xlabel('Age (years)')
plt.ylabel('Sell Price $')
plt.scatter(df['Age(yrs)'], df['Sell Price($)'])
```

```python
# Creating X and Y

x = df[['Mileage','Age(yrs)']]
y = df[['Sell Price($)']]
```

```python
x
```

```python
y
```

```python
# importing the train test split method

from sklearn.model_selection import train_test_split
```

```python
# Splitting the data

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)

# train_test_split() takes independent and dependent variables and size, gives four parameters.
# train_test_split() it chose the random samples everytime we run it 
# if we provide (random_state = 10) it chose the same samples everytime we run it
```

```python
# Finding the length of the training data

len(x_train)
```

```python
# Finding the length of the test data

len(x_test)
```

```python
# Importing the Linear regression model

from sklearn.linear_model import LinearRegression
clf = LinearRegression()
```

```python
# Training the model

clf.fit(x_train,y_train)
```

```python
# Making predictions

clf.predict(x_test)
```

```python
# Actual values

y_test
```

```python
# Calculating score of the test data

clf.score(x_test,y_test)
```

```python
# Calculating score of the train data

clf.score(x_train,y_train)
```
